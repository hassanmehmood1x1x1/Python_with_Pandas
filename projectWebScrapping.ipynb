# extracting seasons too with nested loop

```python
import pandas as pd

root='https://www.football-data.co.uk/mmz4281/'
leagues=['E0','E2','E3']
DataFrames=[]
for league in leagues:
    for season in range(15,21):
        df=pd.read_csv(root+str(season)+str(season+1)+'/'+league+'.csv',encoding='latin1')
        df.insert(1,'Season_Year',str(season))
        DataFrames.append(df)
```

# organizing data with dictionary

```python
dict_countries={'Spanish La Liga':'SP1', 'Spanish Segunda Division':'SP2',
              'German Bundesliga':'D1', 
              'English Premier League':'E0', 'English League 1':'E2', 'English League 2':'E3'
              }
#for league,values in dict_countries.items():
 #   print(values)

# copied code from above as it is only changed "leagues" to "values"

root='https://www.football-data.co.uk/mmz4281/'
#leagues=['E0','E2','E3']           <- now we iterate over dictionary instead of this 
Concat_DataFrames={} #for combined league DataFrames e.g SP1(1516......2021),SP2(1516......2021) an so on

for league,values in dict_countries.items():
    DataFrames=[] # resets for each league 
    for season in range(15,21):
        df=pd.read_csv(root+str(season)+str(season+1)+'/'+values+'.csv',encoding='latin1')
        df.insert(1,'Season_Year',str(season))
        DataFrames.append(df)
        df_concat=pd.concat(DataFrames)
        Concat_DataFrames[league]=df_concat

len(DataFrames)
```

# Creating conditional collumn for more multiple choices with np.select()

```python
conditions=[
    df_laptops['Price_euros']>3000,
    (df_laptops['Price_euros']>2000) &  (df_laptops['Price_euros']<3000), 
    (df_laptops['Price_euros']>800) &  (df_laptops['Price_euros']<2000),
    df_laptops['Price_euros']<800
]
values=['Too Expensive','Expensive','Affordable','Cheap']
Price_tier=np.select(conditions,values,'Unknown')


```

```python
df_laptops['Price_tier']=Price_tier
df_laptops[df_laptops['Price_tier']=='Too Expensive']
```

```python
df_laptops['Price_tier'].value_counts()
```
